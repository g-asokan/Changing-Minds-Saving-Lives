```python
text_topics_csv = "text_topics.csv" 
topics_df = pd.read_csv(text_topics_csv)

text = topics_df["text"]  # open the csv with all the topics
```

```python
classifier = TextClassifier.load('en-sentiment') # TextBlob NLP classifier
str_text = list(map(str, text))
for words in str_text:
    sentence = Sentence(words)
    classifier.predict(sentence)
# print sentence with predicted labels
    print(words, " sentiment is: ", sentence.labels)
```

```python
sid = SentimentIntensityAnalyzer() # flair NLP classifier
with open('./sentiment.csv', "w", newline='') as f:
    writer = csv.writer(f, delimiter=',')
    writer.writerow(['Sentence', 'Flair', 'Vader', 'TextBlob'])  # write the header


# write all three scores (Vader,Flair, and TextBlob for each comment)
for sent in str_text:
    if sent != [] and 'http' not in sent:
        sentiment_single = {}
        sentiment = []
        # Flair
        sentence = Sentence(sent)
        classifier.predict(sentence)
        print('The Sentence is')
        print(sent)
        print('Flair Scores are: ', sentence.labels)
        sentiment_single['Sentence'] = str(sent)
        flair_sign = str(sentence.labels[0]).split(' ')[0]
        flair_score = str(sentence.labels[0]).split(' ')[1].strip(')').strip('(')
        flair_score = float(flair_score)
        if flair_sign == 'NEGATIVE':
            flair_score *= -1
    
        sentiment_single['Flair'] = flair_score # only want polarity score in csv
        
        # Vader
        ss = sid.polarity_scores(sent)
        #for k in sorted(ss):
            #print("Vader Sentiment analysis score is: ")
            #print('{0}: {1}, '.format(k, ss[k]), end='')
            #print()
        sentiment_single['Vader'] = ss['compound']
        print('Vader score is ' + str(tb(sent).sentiment[0]))

        # TextBlob
        print("TextBlob sentiment is: ", tb(sent).sentiment)
        #print(tb(sent).sentiment[0])
        #print(tb(sent).sentiment[1])
        #print(type(tb(sent).sentiment[0]))
        sentiment_single['TextBlob'] = tb(sent).sentiment[0]

        sentiment.append(sentiment_single)
        
        sentiment_df = pd.DataFrame(sentiment)

        sentiment_df.to_csv('./sentiment.csv', mode='a', index=False, header=False)
                        
```

```python
Topic0 = ["circuit", "work", "unemployment", "virus"]
Topic1 = ["america", "trump", "state", "war"]
Topic2 = ["law", "masks", "democrat", "black"]
Topic3 = ["business", "nomral", "money", "shop"]
Topic4 = ["let", "vote", "evil", "time"]

all_topics = [Topic0,Topic1,Topic2,Topic3,Topic4]

data_df = pd.read_csv("sentiment.csv")

#data_sent = data_df["Sentence"]
  
def query_word(topics): # for each topic, the average score is...
    df = data_df[data_df['Sentence'].str.contains(topics, na=False)]
    fscores = df['Flair'].mean()
    vscores = df['Vader'].mean()
    tbscores = df['TextBlob'].mean()
    return [fscores,vscores,tbscores]

def query_topic(list):
    for x in list:
        all_scores = query_word(x)
    all_scores = np.divide(all_scores,len(list))
    return all_scores

for x in all_topics:
    print(query_topic(x))
```
